## Data analysis

### Part 1.1

Part 1.1 is analyzing the group stage data. We will answer several questions:

1. Which team has the highest number of wins?"
2. What is the average number of goals scored per match for each team?
3. What is the average number of goals conceded per match for each team?
4. Which team has the highest goal difference (goals scored minus goals conceded)?
5. Which team has the highest points total?
6. How does the expected goal difference (exp_goal_difference) compare to the actual goal difference (goal_difference) for each team?
7. How do the expected goals scored (exp_goal_scored) and expected goals conceded (exp_goal_conceded) compare to the actual goals scored and conceded for each team?
8. Which team has the highest expected goal difference per 90 minutes played?
9. What is the average rank of teams that scored more and fewer goals than they conceded?
10. What is the average rank of teams that had a positive goal difference?


For answering to this questions we can made some insights and we will understand this data better.

1. Which team has the highest number of wins?"

```python
max_w = df_group_stats['wins'].max()
print("Maximum of win is " + str(max_w))
max_wins = df_group_stats['team'][df_group_stats['wins'] == max_w]
print("Which team has the highest number of wins?")
print(max_wins)
```

**This barplot shows us all teams that wins and devided by groups**

```python
fig = plt.figure(figsize=(18,8))
sns.set_theme(style="whitegrid", palette="pastel")
plt.title('Statistics of win by group')
plt.xticks(rotation=50)
sns.barplot(data=df_group_stats, x="team", y="wins", palette="viridis")
```

 2. What is the average number of goals scored per match for each team?

```python
df_avg_goals = df_group_stats.groupby('team')['goals_scored'].mean().reset_index()

df_avg_goals_per_game = df_avg_goals.assign(goals_scored_per_game = df_avg_goals['goals_scored'] / 3)

print(df_avg_goals_per_game.sort_values(by="goals_scored_per_game",ascending=False))
```

This graph shows all teams goals scored per game

```python
fig = plt.figure(figsize=(18,8))
sns.set_theme(style="whitegrid", palette="pastel")
plt.title('Goal scored per game')
plt.xticks(rotation=50)
sns.barplot(data=df_avg_goals_per_game.sort_values(by="goals_scored_per_game",ascending=False), x="team", y="goals_scored_per_game", palette="viridis")
```

3. What is the average number of goals conceded per match for each team?

```python
avg_goals_conceded = df_group_stats.groupby('team')['goals_against'].mean().reset_index()

avg_goals_conceded_per_game = avg_goals_conceded.assign(goals_against_per_game = avg_goals_conceded['goals_against'] / 3)
print(avg_goals_conceded_per_game.sort_values(by = 'goals_against_per_game',ascending = True))
```

4. Which team has the highest goal difference (goals scored minus goals conceded)?

```python
max_diff = df_group_stats['goal_difference'].max()
print("Maximum goal difference of group stage is "+ str(max_diff))
max_goal_diff = df_group_stats['team'][df_group_stats['goal_difference']==max_diff]
print("Which team has the highest goal difference (goals scored minus goals conceded)?")
print(max_goal_diff)
```

5.  Which team has the highest points total?

```python
max_p = df_group_stats['points'].max()
print("Maximum point in group stage is " + str(max_p))

max_points = df_group_stats['team'][df_group_stats['points']==max_p]
print("Which team has the highest points total?")
print(max_points)
```

6. How does the expected goal difference (exp_goal_difference) compare to the actual goal difference (goal_difference) for each team?

```python
goal_diff_comparison = df_group_stats[['team', 'goal_difference', 'exp_goal_difference']]
print("How does the expected goal difference (exp_goal_difference) compare to the actual goal difference (goal_difference) for each team?")
print(goal_diff_comparison)
```

7. How do the expected goals scored (exp_goal_scored) and expected goals conceded (exp_goal_conceded) compare to the actual goals scored and conceded for each team?


```python
exp_actual_comparison = df_group_stats[['team', 'goals_scored', 'expected_goal_scored', 'goals_against', 'exp_goal_conceded']]
print("How do the expected goals scored (exp_goal_scored) and expected goals conceded (exp_goal_conceded) compare to the actual goals scored and conceded for each team?")
print(exp_actual_comparison)
```

8. Which team has the highest expected goal difference per 90 minutes played?

```python
max_diff_90 = df_group_stats['exp_goal_difference_per_90'].max()
print("Maximum expected goal difference per 90 minutes is "+str(max_diff_90))
max_exp_goal_diff_per_90 = df_group_stats['team'][df_group_stats['exp_goal_difference_per_90'] == max_diff_90]
print(max_exp_goal_diff_per_90)
```

9.  What is the average rank of teams that scored more and fewer goals than they conceded?

```python
# Scored more
df_scored_more = df_group_stats[df_group_stats['goals_scored'] > df_group_stats['goals_against']]
avg_rank_scored_more = df_scored_more['rank'].mean()
print("Average rank of teams that scored more goals than they conceded is "+ str(avg_rank_scored_more))

# Scored fewer or less
df_scored_fewer = df_group_stats[df_group_stats['goals_scored'] < df_group_stats['goals_against']]
avg_rank_scored_fewer = df_scored_fewer['rank'].mean()
print("Average rank of teams that scored fewer goals than they conceded is "+str(avg_rank_scored_fewer))
```

10. What is the average rank of teams that had a positive goal difference?

```python
avg_rank_positive_goal_diff = df_group_stats.loc[df_group_stats['goal_difference'] > 0]['rank'].mean()
print("Average rank of teams that had a positive goal difference is "+str(avg_rank_positive_goal_diff))
```
